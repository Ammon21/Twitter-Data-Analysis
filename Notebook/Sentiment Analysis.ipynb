```python
cleanTweet = df[["original_text", "polarity"]]
cleanTweet
```

```python
def pol_test(p):
    """
    converts polarity into sentiment category
    """
    if p > 0:
        return "positive"
    elif p < 0:
        return "negative"
    else:
        return "neutral"
```

```python
cleanTweet["score"] = cleanTweet["polarity"].apply(pol_test)
cleanTweet
```

```python
sentiment = cleanTweet['score'].value_counts()
sentiment
```

```python
cleanTweet = cleanTweet[cleanTweet['score'] != 'neutral']
cleanTweet
```

```python
# add scoremap column
#authored by
def scoremap(score):
    """
    maps sentiment score to 0 and 1
    """
    if score == "positive":
        return 1
    else:
        return 0
```

```python
cleanTweet['scoremap'] = cleanTweet["score"].map(scoremap)
cleanTweet
```

```python
from sklearn.linear_model import SGDClassifier
from sklearn.model_selection import train_test_split
from scipy.sparse import csr_matrix
import numpy as np
```

```python
def train_and_show_scores(X: csr_matrix, y: np.array, title: str) -> None:
    X_train, X_test, y_train, y_test = train_test_split(
        X, y, train_size=0.75, stratify=y
    )

    clf = SGDClassifier()
    clf.fit(X_train, y_train)
    train_score = clf.score(X_train, y_train)
    test_score = clf.score(X_test, y_test)
    print(f'{title}\nTrain score: {round(train_score, 2)} ; Test score: {round(test_score, 2)}\n')
```

```python
train_and_show_scores(X_trigram_vectorizer, y.values, 'Trigram Counts')
```

```python
nltk.download('punkt')
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize

class PrepareData:
  def __init__(self,df):
    self.df=df
    
  def preprocess_data(self):
    df = self.df.loc[self.df['language'] =="en"]

    
    #text Preprocessing
    df['original_text']=   df['original_text'].astype(str)
    df['original_text'] =  df['original_text'].apply(lambda x: x.lower())
    df['original_text']=   df['original_text'].apply(lambda x: x.translate(str.maketrans(' ', ' ', string.punctuation)))
    
    #Converting tweets to list of words For feature engineering
    sentence_list = [tweet for tweet in df['original_text']]
    word_list = [sent.split() for sent in sentence_list]
    
    
    
    

    #Create dictionary which contains Id and word 
    word_to_id = corpora.Dictionary(word_list)
    corpus_1= [word_to_id.doc2bow(tweet) for tweet in word_list]



    
    return word_list, word_to_id, corpus_1
```

```python
nltk.download('punkt')
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize

stop_words = set(stopwords.words('english'))
 
word_tokens = word_tokenize(''.join(word_list))
 
word_list = [w for w in word_tokens if not w.lower() in stop_words]
 
word_list = []
 
for w in word_tokens:
    if w not in stop_words:
        word_list.append(w)
 
print(word_list)
```

```python
PrepareData_obj=PrepareData(df)
word_list ,id2word,corpus=PrepareData_obj.preprocess_data()
```

# Topic Modeling using Latent Dirichlet Allocation


```python
lda_model = gensim.models.ldamodel.LdaModel(corpus,
                                           id2word=id2word,
                                           num_topics=5, 
                                           random_state=100,
                                           update_every=1,
                                           chunksize=100,
                                           passes=10,
                                           alpha='auto',
                                           per_word_topics=True)
```

```python
print(lda_model.show_topics(formatted=False))
```

```python
print('\nPerplexity: ', lda_model.log_perplexity(corpus))  
doc_lda = lda_model[corpus]


# Compute Coherence Score
coherence_model_lda = CoherenceModel(model=lda_model, texts=word_list, dictionary=id2word, coherence='c_v')
coherence_lda = coherence_model_lda.get_coherence()
print('\n Ldamodel Coherence Score/Accuracy on Tweets: ', coherence_lda)
```
